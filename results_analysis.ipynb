```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# ['DSI','LIB','TUT1','TUT2','TUT3','Uji']
dataset_name = 'Uji'
weighted_knn = 0
rsl_knn_path = 'results/knn_' + dataset_name+'.csv'
rsl_gknn_path = 'results/gknn_' + dataset_name+'.csv'
```

```python
rsl_knn = pd.read_csv(rsl_knn_path)
rsl_gknn = pd.read_csv(rsl_gknn_path)
```

```python
knn_mae = rsl_knn[rsl_knn['weighted knn']==weighted_knn][['k','MAE']].values
gknn_mae = rsl_gknn[rsl_gknn['weighted knn']==weighted_knn][['n','k','MAE']].values
```

```python
a = np.asarray(np.unique(gknn_mae[:,0]))
b = np.asarray(np.unique(gknn_mae[:,1]))
```

```python
plt.figure()

for i in a:
    data_to_plot = gknn_mae[gknn_mae[:,0]==i,-2:]
    plt.plot(data_to_plot[:,0], data_to_plot[:,1], marker='o', linestyle='-',label=f'gknn:n={i}')

plt.plot(knn_mae[:,0], knn_mae[:,1], marker='>', linestyle='-', label='KNN',color = 'black')
ttl = dataset_name + ': k vs MAE'
plt.title(ttl)
plt.xlabel('k values')
plt.ylabel('MAE(m)')
plt.grid(True)
plt.figlegend(loc='upper left', bbox_to_anchor=(1,1))
plt.show()
```

```python
knn_80p = rsl_knn[rsl_knn['weighted knn']==weighted_knn][['k','80 percentile error']].values
gknn_80p = rsl_gknn[rsl_gknn['weighted knn']==weighted_knn][['n','k','80 percentile error']].values
```

```python
plt.figure()

for i in a:
    data_to_plot = gknn_80p[gknn_80p[:,0]==i,-2:]
    plt.plot(data_to_plot[:,0], data_to_plot[:,1], marker='o', linestyle='-',label=f'gknn:n={i}')

plt.plot(knn_80p[:,0], knn_80p[:,1], marker='>', linestyle='-', label='KNN',color = 'black')
ttl = dataset_name + ': k vs 80% error'
plt.title(ttl)
plt.xlabel('k values')
plt.ylabel('80 Percentile Error(m)')
plt.grid(True)
plt.figlegend(loc='upper left', bbox_to_anchor=(1,1))
plt.show()
```
